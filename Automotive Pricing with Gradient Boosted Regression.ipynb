## Predicting Vehicle Value with a Gradient Boosting Regressor

A vehicle valuation company wants to build a model that can predict the price of a car based on its features: year, mileage, trim package, etc. A database of car features and their prices is provided.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import scale
from sklearn.ensemble import GradientBoostingRegressor as GBR
from sklearn.grid_search import GridSearchCV
```

Import the data. Preview the dataframe.

```python
df = pd.read_csv('Cars.csv')
print(df.shape)
print(df.head())
```

Basic statistical analysis of numeric features.

```python
df.describe()
```

```python
df.info()
```

Because we are going to use dummy variables, let's take a closer look at the number of unique categories in each of the categorical columns.

```python
df_obj = df.select_dtypes(include=['O'])
print("Number of unique categories in feature:")
for col in df_obj:
    print("{}:  ".format(col),df_obj[col].nunique())
```

To facilitate feature selection, we examine violin plots of categorical features with respect to price. Similar distributions in a feature means that extracting signal from that feature will be more difficult. 

```python
for col in df_obj:
    fig, ax = plt.subplots(figsize=(15,df_obj[col].nunique()))
    sns.violinplot(df['price'],df_obj[col])
```

We set target and feature data. We drop the 'state' column as it is noisy and hightly correlated with the 'region' column.

```python
y = df['price']
X = df.drop(['price','state','Index'],axis=1)
```

Preprocess our feature data by scaling numeric features and getting dummy variables for our categorical features.

```python
X_num = X.select_dtypes(include=['int64','float64'])
X_cat = X.select_dtypes(include=['O'])

X_num = pd.DataFrame(scale(X_num))
X_cat = pd.get_dummies(X_cat)

X = pd.concat([X_num,X_cat],axis=1)
```

We train our model with a Gradient Boosted Regressor, using 5-fold cross validation, and search over a selection of hyperparameters using GridSearchCV. 

Note: Because we know that the learning rate and the number of estimators should have a roughly inverse relationship, we expect combinations of low number estimators with low learning rates to have poor accuracy. Additionally, we will examine a heatmap of accuracy vs. parameter selection to confirm our best parameters.

```python
params = {'n_estimators':[50,100,500,1000],'learning_rate':[.01,.1,1]}
clf = GridSearchCV(GBR(), params, cv=5, verbose=1)
clf.fit(X,y)
print(clf.best_score_)
print(clf.best_params_)
```

```python

scores = [x[1] for x in clf.grid_scores_]
scores = np.array(scores).reshape(len(params['learning_rate']), len(params['n_estimators']))

plt.figure(figsize=(12, 8))
plt.imshow(scores, interpolation='nearest', cmap=plt.cm.hot, vmin=.55, vmax=.98)
plt.xlabel('n_estimators')
plt.ylabel('learning_rate')
plt.colorbar()
plt.xticks(np.arange(len(params['n_estimators'])), params['n_estimators'], rotation=45)
plt.yticks(np.arange(len(params['learning_rate'])), params['learning_rate'])
plt.title('Validation accuracy')
plt.show()
```

We can imagine the utility of an accurate model for any automotive company.
